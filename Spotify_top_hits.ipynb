```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv(f'/content/songs_normalize.csv')
```

```python
df.head()
```

```python
df.shape
```

```python
# remove any instance after "," in the entries stored under "genre"

df['genre'] = df['genre'].str.split(',', expand=True)[0]
```

```python
df.head()
```

```python
# get unique values with their counts in the artist column

df['artist'].value_counts()
```

```python
# drop artist and song from data

df.drop(['artist', 'song'], axis=1, inplace=True)
```

```python
df.head()
```

```python
# encode genre without creating dummy variables

from sklearn.preprocessing import LabelEncoder

label_encoder = LabelEncoder()
df['genre'] = label_encoder.fit_transform(df['genre'])
```

```python
# encode explicit without creating dummy variables

df['explicit'] = label_encoder.fit_transform(df['explicit'])
```

```python
df.head()
```

```python
#  function to create distribution plots of all columns

def distribution_plots(df):
  for column in df.columns:
    sns.distplot(df[column])
    plt.xlabel(column)
    plt.ylabel('Frequency')
    plt.title(f'Distribution of {column}')
    plt.show()
```

```python
distribution_plots(df)
```

```python
# which genre is the loudest?

# Group the data by genre and calculate the mean loudness for each genre
genre_loudness = df.groupby('genre')['loudness'].mean()

# Sort the genres by loudness in descending order
loudest_genre = genre_loudness.sort_values(ascending=False).index[0]

# Print the loudest genre
print("The loudest genre is:", loudest_genre)
```

```python
# check which genre was encoded as 7

print(label_encoder.inverse_transform([7]))
```

```python
# which genre is the most danceble?

# group data by danceability and calculate mean
genre_dance = df.groupby('genre')['danceability'].mean()

# sort genres by danceability in descending order
most_danceable = genre_dance.sort_values(ascending = False).index[0]

# print the most danceable genre
print("The most danceable genre is: ", most_danceable)
```

```python
# which genre is the most explicit?

# group data by explicit rating and calculate mean
genre_explicit = df.groupby('genre')['explicit'].mean()

# sort genres by explicity rating in descending order
explicit_music = genre_explicit.sort_values(ascending = False).index[0]

# print the code of the most explicit genre
print("The most explicit genre of music is: ", explicit_music)
```

```python
# function which charts the most popular genre, most explicit genre and genre with the most danceability

def plot_genre_stats(df):
  # Calculate genre popularity
  genre_counts = df['genre'].value_counts()

  # Get the most popular genre
  most_popular_genre = genre_counts.index[0]

  # Calculate genre explicitness
  genre_explicitness = df.groupby('genre')['explicit'].mean()

  # Get the most explicit genre
  most_explicit_genre = genre_explicitness.idxmax()

  # Calculate genre danceability
  genre_danceability = df.groupby('genre')['danceability'].mean()

  # Get the most danceable genre
  most_danceable_genre = genre_danceability.idxmax()

  # Create a bar chart of genre popularity
  plt.figure(figsize=(10, 6))
  sns.barplot(x=genre_counts.index, y=genre_counts.values)
  plt.xlabel('Genre')
  plt.ylabel('Count')
  plt.title('Genre Popularity')
  plt.xticks(rotation=90)
  plt.show()

  # Create a bar chart of genre explicitness
  plt.figure(figsize=(10, 6))
  sns.barplot(x=genre_explicitness.index, y=genre_explicitness.values)
  plt.xlabel('Genre')
  plt.ylabel('Explicitness')
  plt.title('Genre Explicitness')
  plt.xticks(rotation=90)
  plt.show()

  # Create a bar chart of genre danceability
  plt.figure(figsize=(10, 6))
  sns.barplot(x=genre_danceability.index, y=genre_danceability.values)
  plt.xlabel('Genre')
  plt.ylabel('Danceability')
  plt.title('Genre Danceability')
  plt.xticks(rotation=90)
  plt.show()

  # Print the most popular, explicit, and danceable genres
  print('Most Popular Genre:', most_popular_genre)
  print('Most Explicit Genre:', most_explicit_genre)
  print('Most Danceable Genre:', most_danceable_genre)

# Call the function
plot_genre_stats(df)
```
